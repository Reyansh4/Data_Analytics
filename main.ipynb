Lets Do preprocessing and make a new csv for Data Visualization in Tableau.

```python
data = data.dropna()
data
```

```python
#lets see for null values in the pandas again
print(data.isna().sum())
```

```python
#lets drop duplicate values in the pandas
data = data.drop_duplicates()
data
```

```python
msno.matrix(data)
```

```python
crop_counts = data.groupby('Crop')['Crop_Year'].nunique()

crops_harvested_every_year = crop_counts[crop_counts == data['Crop_Year'].nunique()].index

new_data = data[data['Crop'].isin(crops_harvested_every_year)]
```

```python
print("The shape of the datasets is : ", new_data.shape)
print("The no of rows in the data are : ", new_data.shape[0])
print("The no of columns in the dataset are : ", new_data.shape[1])
```

```python
new_data.to_csv("result.csv", index=False)
```
