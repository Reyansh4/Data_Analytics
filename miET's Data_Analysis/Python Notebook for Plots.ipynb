```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Gene_details_T_vs_C.csv")
data
```

```python
count = 0
for i in data['logFC']:
  if (i<= 10.0 and i >= -10.0):
    #print(i)
    count= count + 1

print("The no of Genes out the range of -10 and 10 logFC are : ",len(data) - count)

print("The Total no of unique genes in the experiment are : ",len(data['Gene_Name'].unique()))
```

```python
def logfcfunction(gene_name):
  count = 0
  for i in data['Gene_Name']:
    count= count + 1
    if (i == gene_name):
      return(data['logFC'][count])

X = ['Padi3','Ldhb','Ero1a','Mpo','Ldhc','Tkt','Vdac2','Eif5a','Padi1','Padi4','Ero1b','Padi2','Lcp1','Slc25a31','Hspa9','Anxa2','Cat','Eif5a2','Slc25a5','Padi3','Eno1','Sod1','Ptma','Actn4','Prkca','Tuba1b','Eef2','Vdac3','Elane','Ak2','Mdh2','Calu','Hsp90ab1','Hspa5','P4hb','Cct7']
Y = [logfcfunction(i) for i in X]
```

```python
plt.figure(figsize = (20,10))
plt.bar(X,Y)
plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel('Gene_Name')
plt.xticks(rotation = 90)
plt.ylabel('logFC values')
```

```python
X_new = ['Padi3','Ldhb','Ero1a','Tkt','Vdac2','Eif5a','Padi1','Ero1b','Padi2','Lcp1','Hspa9','Anxa2','Cat','Eif5a2','Slc25a5','Padi3','Eno1','Sod1','Ptma','Actn4','Prkca','Tuba1b','Eef2','Vdac3','Ak2','Mdh2','Calu','Hsp90ab1','Hspa5','P4hb','Cct7']
Y_new = [logfcfunction(i) for i in X_new]

plt.figure(figsize = (20,10))
plt.bar(X_new,Y_new)
plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel('Gene_Name')
plt.xticks(rotation = 90)
plt.ylabel('logFC values')
```

```python
def LogFC(external_gene_name):
  count = 0
  for i in new_data['external_gene_name']:
    count= count + 1
    if (i == external_gene_name):
      return(new_data['Log FC'][count])
```

```python
X_std = ['B020010K11Rik','Rian','Gm20721','Celsr1','Ube2o','Srp54b','Dtl','Zfp335','Tomm6','Pycrl','H2ac14-ps','Bcat2','Sf3b4','Per1','Derpc','Gm26448','Rnu1b6','Gm26232','Ptgs2','Padi3','Ldhb','Ero1a','Mpo','Pdcd6ip','Ltf','Myh9','Ldhc','Tkt',
'Vdac2','Eif5a','Padi1','Padi4','Ero1b','Padi2','Lcp1','Slc25a31','Hspa9','Anxa2','Cat','Eif5a2','Padi6','Slc25a5','Padi3','Eno1','Sod1','Ptma','Actn4','Prkca','Tuba1b','Eef2','Vdac3','Elane','Ak2','Mdh2','Calu','Lyz1','Lyz2','Hsp90ab1',
         'Hspa5','P4hb','Cct7','Gm24830']

Y_std = [LogFC(i) for i in X_std]

change = 0
for i in range(0,len(Y_std)):
  if (Y_std[i] == 'Nan'):
    change = change + 1
    Y_std[i] = 0

mean_value = np.mean(Y_std)
standard_deviation = np.std(Y_std, ddof=1)
sample_size = len(Y_std)
standard_error_mean = standard_deviation / np.sqrt(sample_size)

print("Mean:", mean_value)
print("Standard Deviation:", standard_deviation)
print("Standard Error of the Mean (SEM):", standard_error_mean)
print("The Total no of genes in the required dataste are : ", len(X_std))
print("The Total no of genes where FC values cannot be defined are : ",change)
```

```python
plt.figure(figsize = (30,10))
plt.bar(X_std,Y_std, yerr = standard_error_mean, capsize = 5)
plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel('Gene_Name')
plt.xticks(rotation = 90)
plt.ylabel('logFC values')
```
